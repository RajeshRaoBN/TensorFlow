```python
# Import the necessary packages and modules
import matplotlib.pyplot as plt
import numpy as np

# Prepare the data
x = np.linspace(0, 10, 100)

# Plot the data
plt.plot(x, x, label='linear')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# X-Axis
x = np.linspace(0, 5, 5)

# Plot the square
plt.plot(x, x**2, label='square')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# X-Axis
x = np.linspace(1,10,10)

# Generate plot
plt.plot(x, x**3, color='red', label="Cube Function")
plt.plot(x, x**2, color='blue', label="Square Function")
plt.plot(x, x, color='green', label="Linear Function")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
ax.set_xlim(0.5, 4.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1, 2, 3, 4], [30, 25, 20, 15], color = 'green', label='Linegraph')
ax.scatter([1, 2, 3, 4], [10, 15, 25, 30], color = 'darkblue', label='Scatterplot', marker = '^')
plt.legend()
plt.show()
```
